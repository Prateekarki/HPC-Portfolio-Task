```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#original data dataset
data1 = pd.read_csv('csv files/linear.csv')
data1.head()
```

```python
#scatter plot of original data
plt.scatter(data1.x, data1.y, c='grey')
plt.title("Original Data Plot")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```

```python
#first guess import
first = pd.read_csv('csv files/firstguess.csv')
first.head()
```

```python
#first guess plot
plt.scatter(data1.x, data1.y, c='grey')
plt.plot(first.x, first.y, c='r')
plt.title("First guess Data Plot")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```

```python
#second guess import
second = pd.read_csv('csv files/secondguess.csv')
second.head()
```

```python
#second guess plot
plt.scatter(data1.x, data1.y, c='grey')
plt.plot(second.x, second.y, c='b')
plt.title("Second Guess Data Plot")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```

```python
#thirst guess import
third = pd.read_csv('csv files/thirdguess.csv')
third.head()
```

```python
#third guess plot
plt.scatter(data1.x, data1.y, c='grey')
plt.plot(third.x, third.y, c='y')
plt.title("Third Guess Data Plot")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```

```python
#all three guesses
plt.scatter(data1.x, data1.y, c= 'grey')
plt.plot(first.x, first.y, c='r')
plt.plot(second.x, second.y, c='b')
plt.plot(third.x, third.y, c='y')
plt.title("All three guesses Plot")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```

```python
best = pd.read_csv('csv files/bestguess.csv')
best.head()
```

```python
#best guess plot
plt.scatter(data1.x, data1.y, c='grey')
plt.plot(best.x, best.y, c='w')
plt.title("Original Data Plot")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```

```python
#all guesses with best guess plot
plt.scatter(data1.x, data1.y, c='grey')
plt.plot(first.x, first.y, c='r')
plt.plot(second.x, second.y, c='b')
plt.plot(third.x, third.y, c='y')
plt.plot(best.x, best.y, c='w')
plt.title("3 Gusses with Best M & C")
plt.xlabel("X")
plt.ylabel("y = f(x)")
plt.grid()
plt.show()
```
